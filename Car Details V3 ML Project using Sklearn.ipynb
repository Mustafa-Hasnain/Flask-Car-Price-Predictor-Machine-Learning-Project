```python
#iloc is usually used to get the index location from the table
iloc = final_dataset.iloc() #example
print(iloc[0:5])

#Dependent Feature Y
y = final_dataset.iloc[:,0] #This Means the very First Column

#Independent Features x
X = final_dataset.iloc[:,1:] #This Means The Columns after the First Column
```

```python
#importing sklearn
from sklearn.ensemble import ExtraTreesRegressor
######Feature Importance#######
model = ExtraTreesRegressor()  #This Regression Algorithm will help you in finding out the best features
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
#Plotting the Graph For better visualization
feat_importances = pd.Series(model.feature_importances_,index=X.columns)
feat_importances.nlargest(5).plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#Here We are using Random Forest Regression
from sklearn.ensemble import RandomForestRegressor
rf_random = RandomForestRegressor()
```

```python
#Randomized Search CV

#Setting up the hyper parameters
n_estimators = [int(x) for x in np.linspace(start=100, stop=1000, num=12)]

#No of features Consider at every split
max_features = ['auto', 'sqrt']

#Maximum no of levels in a tree
max_dept = [int(x) for x in np.linspace(5, 30, num=6)]

#Minimun number of samples required to split a node
min_samples_split = [2,5,10,15,100]

#Minimum number of samples required at each Leaf node
min_samples_leaf = [1,2,5,10]
```

```python
#To get the Best Parameters we will use Randomized Search CV

##### Here CV stands For CROSS VALIDATION########
from sklearn.model_selection import RandomizedSearchCV

#You can also Used GridSearch CV instead of Random Search But Random Search is much faster
#from sklearn.model_selection import GridSearchCV
```

```python
#Create Grid For Randomized Search CV
random_grid = {'n_estimators':n_estimators,
'max_features':max_features,
'min_samples_split':min_samples_split,
'min_samples_leaf':min_samples_leaf}

print(random_grid)
```

```python
rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator=rf, param_distributions=random_grid, scoring='neg_mean_squared_error',
                               n_iter=10, cv=5, random_state=42, n_jobs=1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
#Getting the Best Parameters
rf_random.best_params_
```

```python
#Getting the Score for the model
rf_random.best_score_
```

```python
#Applying the test data on the model
predict = rf_random.predict(X_test)
```

```python
sns.displot(y_test-predict)
```

```python
#Plotting the Predictions
sns.distplot(y_test-predict)
```

```python
#Getting the Cluster Data/Scatter Data
sns.scatterplot(y_test,predict)
```

```python
#Getting the Error using sklearn.metrices
from sklearn import metrics
print("The Mean Absolute Error is: ",metrics.mean_absolute_error(y_test, predict))
print("The Mean Squared Error is: ", metrics.mean_squared_error(y_test, predict))

#Just Take the Sqrt of Mean Squared Error
print("The Root Mean Squared Error is: ",np.sqrt(metrics.mean_squared_error(y_test, predict)))
```
